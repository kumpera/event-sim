```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import os
```

```python
comp = pd.read_csv('win-compress.csv')
decomp = pd.read_csv('win-decompress.csv')
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.set_xlabel('compression level')
ax.set_ylabel('compression time in secs')

def plot(method):
    z = comp[comp.method == method]
    plt.errorbar(z.level, z['mean'] / 1_000, yerr=z['std'] / 1_000, label=method)

for m in list(comp.method.unique()):
    plot(m)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.set_xlabel('compression level')
ax.set_ylabel('compression ratio')

def plot_ratio(method):
    z = comp[comp.method == method]
    plt.plot(z.level, z['ratio'], label=method)

for m in list(comp.method.unique()):
    plot_ratio(m)

plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.set_xlabel('compression level')
ax.set_ylabel('decompression time in secs')

def plot(method):
    z = decomp[decomp.method == method]
    plt.errorbar(z.level, z['mean'] / 1_000, yerr=z['std'] / 1_000,label=method)

for m in list(decomp.method.unique()):
    plot(m)
plt.legend()
```
